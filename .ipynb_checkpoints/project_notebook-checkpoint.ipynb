```python
#calcuar dias q tarda en recibir el dinero desde petición
# Eliminar columna sent_at recovery_status reco_creation
# frecuencia, vemos por user id cuales han repetido
# Tasa incidentes con el csv de fees
# calcular tasa cuentas canceladas
```

```python
df = (
    df.drop(columns=['updated_at', 'moderated_at', 'send_at'])
    .rename(columns={"id": "cash_request_id"})
)
df
```

```python
df_combined = pd.merge(df, df_fees, on='cash_request_id', how='outer')
df_combined
```

```python
df_combined.info()
```

```python
df_combined = (df_combined
    .drop(columns=['created_at_y', 'status_y'])
    .rename(columns={"status_x": "status", "created_at_x": "created_at", "id": "fees_id"}))
df_combined
```

```python
df_combined = df_combined.dropna(subset=['cash_request_id'])
df_combined
```

```python
group_df = df_combined.groupby("cash_request_id").count()
group_df.sort_values(by="cash_request_id", ascending=True)
```
