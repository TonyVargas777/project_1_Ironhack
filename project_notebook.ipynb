```python
df_cash = (
    df_cash.drop(columns=['updated_at', 'moderated_at', 'send_at'])
    .rename(columns={"id": "cash_request_id"})
)
df_cash
```

```python
pd.set_option('display.max_columns', None)
df_combined = pd.merge(df_cash, df_fees, on='cash_request_id', how='outer')
df_combined
```

```python
df_combined.info()
```

```python
df_combined = (df_combined
    .drop(columns=['created_at_y', 'status_y', "reason", "updated_at", "from_date", "to_date", "recovery_status", "reco_last_update", "reco_creation"])
    .rename(columns={"status_x": "status", "created_at_x": "created_at", "id": "fees_id", "reimbursement_date": "expected_date", "type": "fee_type", "category": "incident_category", "total_amount": "fee_amount"}))
df_combined
```
